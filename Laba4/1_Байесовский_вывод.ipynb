# Задача 1

Пусть $X_1, X_2, \ldots, X_n$ — выборка из экспоненциального распределения с параметром $\lambda$. Найти оценку максимального правдоподобия параметра $\lambda$, сравнить ее с байесовской оценкой (MAP и математическое ожидание апостреорного распределения), подобрав сопряженное распределение. Сравнить полученные байесовские оценки с оценкой MLE. Найти предсказательное распределение

## **Решение**

Пусть наблюдается выборка  
$X_1, X_2, \dots, X_n$ — независимые случайные величины, и при фиксированном параметре $\lambda$ выполняется

$$
X_i \mid \lambda \sim \mathrm{Exp}(\lambda)
$$

Плотность имеет вид

$$
f(x \mid \lambda) = \lambda e^{-\lambda x}, \quad x \ge 0
$$

Сумма наблюдений

$$
S = \sum_{i=1}^{n} X_i
$$


### **Функция правдоподобия**

Совместная плотность выборки равна

$$
L(\lambda \mid x_1,\dots,x_n)
= \lambda^n e^{-\lambda S}
$$


### **Гамма-априор**

Предположим, что параметр имеет априорное распределение

$$
\lambda \sim \mathrm{Gamma}(\alpha, \beta),
$$

с плотностью

$$
\pi(\lambda)
= \frac{\beta^{\alpha}}{\Gamma(\alpha)} \lambda^{\alpha - 1} e^{-\beta \lambda}
$$
## **Апостериорное распределение**

По формуле Байеса

$$
\pi(\lambda \mid x_1,\dots,x_n)
\propto
\lambda^n e^{-\lambda S}
\cdot
\lambda^{\alpha - 1} e^{-\beta \lambda}
=
\lambda^{\alpha+n-1} e^{-(\beta+S)\lambda}
$$

Получаем

$$
\lambda \mid x_1,\dots,x_n
\sim
\mathrm{Gamma}(\alpha + n,\ \beta + S)
$$


## **MLE и Байесовская оценка**

Оценка максимального правдоподобия

$$
\hat{\lambda}_{\mathrm{MLE}} = \frac{n}{S}
$$

Байесовская оценка

$$
\hat{\lambda}_{\mathrm{Bayes}}
= \frac{\alpha + n}{\beta + S}
$$

## **Предсказательное распределение**

Пусть новое наблюдение

$$
X_{n+1} \mid \lambda \sim \mathrm{Exp}(\lambda)
$$

Тогда предсказательная плотность

$$
p(x_{n+1} \mid X)
= \alpha' (\beta')^{\alpha'}(\beta' + x_{n+1})^{-(\alpha' + 1)},
$$

где

$$
\alpha' = \alpha + n, \quad \beta' = \beta + S
$$

# Задача 2

**Мультиномиальное распределение**

Пусть проводится серия из $n$ испытаний и в результате каждого испытания происходит ровно одно событие из набора $A_1, A_2, \dots, A_m$, причем вероятности этих событий равны соответственно $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, причем
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

Тогда совместное распределение величин $X_1, X_2, \dots, X_m$, где $X_k$ — число наступлений события $A_k$ в серии из $n$ испытаний, задается вероятностями

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

где $n_1, n_2, \dots, n_m$ — произвольный набор целых неотрицательных чисел, таких что

$$\sum_{i=1}^m n_i = n.$$

Произведите байесовский вывод для мультиномиального распределения: найдите апостериорное распределение, используя в качестве сопоряженного распределения к правдоподобию [распределение Дирихле](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), найдите предсказательное распределение. Объясните результат.

## **Решение**

### Априорное распределение (Дирихле)

Введём априорное распределение на вектор вероятностей $\mathbf{p}$ — **распределение Дирихле** с параметрами
$$
\boldsymbol{\alpha} = (\alpha_1,\dots,\alpha_m), \quad \alpha_i > 0
$$

Запишем
$$
\mathbf{p} \sim \mathrm{Dir}(\alpha_1,\dots,\alpha_m)
$$

Плотность распределения Дирихле
$$
\pi(\mathbf{p})
=
\frac{1}{B(\boldsymbol{\alpha})}
\prod_{i=1}^{m} p_i^{\alpha_i - 1},
\quad
p_i \ge 0, \quad \sum_{i=1}^{m} p_i = 1.
$$

Здесь
$$
B(\boldsymbol{\alpha})
=
\frac{\prod_{i=1}^{m} \Gamma(\alpha_i)}{\Gamma\!\left(\sum_{i=1}^{m} \alpha_i\right)}
$$
— многомерная бета-функция. Часто $\alpha_i$ интерпретируют как **псевдонаблюдения** (фиктивные счётчики) для каждого исхода.

### Правдоподобие

Правдоподобие при наблюдённом векторе счётчиков $\mathbf{n} = (n_1,\dots,n_m)$
$$
L(\mathbf{p} \mid \mathbf{n})
=
\mathsf{P}(\mathbf{X} = \mathbf{n} \mid \mathbf{p})
=
\frac{n!}{n_1! \dots n_m!}
\prod_{i=1}^{m} p_i^{n_i}
$$

С точностью до константы (не зависящей от $\mathbf{p}$)
$$
L(\mathbf{p} \mid \mathbf{n}) \propto \prod_{i=1}^{m} p_i^{n_i}
$$

### Апостериорное распределение

По формуле Байеса
$$
\pi(\mathbf{p} \mid \mathbf{n})
\propto
L(\mathbf{p} \mid \mathbf{n}) \, \pi(\mathbf{p})
$$

Подставляя выражения для правдоподобия и априора, получаем
$$
\pi(\mathbf{p} \mid \mathbf{n})
\propto
\left( \prod_{i=1}^{m} p_i^{n_i} \right)
\left( \prod_{i=1}^{m} p_i^{\alpha_i - 1} \right)
=
\prod_{i=1}^{m} p_i^{\alpha_i + n_i - 1}
$$

Это снова плотность распределения Дирихле, поэтому апостериорное распределение имеет вид
$$
\mathbf{p} \mid \mathbf{n}
\sim
\mathrm{Dir}(\alpha_1 + n_1, \dots, \alpha_m + n_m)
$$

Если обозначить
$$
\alpha_0 = \sum_{i=1}^{m} \alpha_i,
\quad
n = \sum_{i=1}^{m} n_i,
$$
то апостериорный вектор параметров — это просто **сумма априорных псевдонаблюдений и реальных счётчиков**

### Байесовская оценка для $\mathbf{p}$

При квадратичной функции потерь байесовская оценка — это математическое ожидание апостериорного распределения. Для распределения Дирихле верно:

$$
\mathbb{E}[p_i \mid \mathbf{n}]
=
\frac{\alpha_i + n_i}{\alpha_0 + n}, \quad i = 1,\dots,m
$$

$$
\hat{p}_i^{\text{Bayes}}
=
\frac{\alpha_i + n_i}{\alpha_0 + n}
$$

Особый случай — **сглаживание Лапласа**: если $\alpha_i = 1$ для всех $i$, то
$$
\hat{p}_i^{\text{Bayes}}
=
\frac{n_i + 1}{n + m}
$$

Здесь к каждому счётчику добавляется 1, а к общему числу испытаний — $m$.


### Предсказательное распределение

Теперь найдём **предсказательное распределение** для результата следующего испытания $Y$.

Пусть $Y$ принимает значение $k \in \{1,\dots,m\}$, если в следующем испытании наступило событие $A_k$.

По определению предсказательной вероятности
$$
\mathsf{P}(Y = k \mid \mathbf{n})
=
\int \mathsf{P}(Y = k \mid \mathbf{p}) \, \pi(\mathbf{p} \mid \mathbf{n}) \, d\mathbf{p}
$$

Условно при фиксированном $\mathbf{p}$
$$
\mathsf{P}(Y = k \mid \mathbf{p}) = p_k
$$

Тогда
$$
\mathsf{P}(Y = k \mid \mathbf{n})
=
\int p_k \, \pi(\mathbf{p} \mid \mathbf{n}) \, d\mathbf{p}
=
\mathbb{E}[p_k \mid \mathbf{n}]
$$


Следовательно,
$$
\boxed{
\mathsf{P}(Y = k \mid \mathbf{n})
=
\frac{\alpha_k + n_k}{\alpha_0 + n}
}
$$

— это и есть **предсказательное распределение** для следующего исхода.

### Суть результата

- **Дирихле — сопряжённый априор** к мультиномиальному распределению, поэтому апостериор тоже Дирихле
- **Параметры апостериора = априорные псевдонаблюдения + реальные наблюдения**  
- **Предсказательная вероятность для следующего исхода — это сглаженная частота:**
  - учитывает и данные, и априор
  - никогда не даёт нулевых вероятностей
  - хорошо работает, когда выборка мала или некоторые исходы не встречались
